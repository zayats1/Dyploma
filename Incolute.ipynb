```python
import math
import numpy as np
import matplotlib.pyplot as plt



t =  np.linspace(0,math.pi/2)
r = 10
D = 2*r


x= r *(np.cos(t)+t*np.sin(t))
y = r* (np.sin(t) - t* np.cos(t)) 

fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

circle1 = plt.Circle((0, 0), r, color='g',fill=False)
ax.add_patch(circle1)

step = 5
ax.set(xlim=(-D,D), xticks=np.arange(-D,D,step),
       ylim=(-D,D), yticks=np.arange(-D,D,step))

plt.show()
```
